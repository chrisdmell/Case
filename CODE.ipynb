### 1. Create a table from Data 1 where MPI of Urban and Rural are greater than It’s Average.


```python
import pandas as pd
```

```python
df1 = pd.read_csv ('Data1.csv')
df2 = pd.read_csv ('Data2.csv')
```

```python
df1.isnull().values.any()
```

```python
df2.isnull().values.any()
```

```python
df3=df1[df1['MPI Urban'] > df1['MPI Urban']. mean()]
df4=df3[df3['MPI Rural'] > df3['MPI Rural']. mean()]
print ('Table where MPI of Urban and Rural are greater than It’s Average')
df4
```

```python
df4.to_csv('answer1.csv')
```

### 2. Get the Count of Rows where any of the line Items is not having the Value (from Data1 and Data2)


```python
print ('The no of rows in Data1.csv and Data2.csv where rows have null value are {} and {} respectively'.format(df1.isnull().any().sum(),df2.isnull().any().sum()))
```

### 4. Create a Table where Intensity of Rural Deprivation is less than the average(Data1)

### 5. Analyze the Data from both the Datasets and Try to Get the Country level Data map to its Sub-national region data.

```python
df2.rename(columns={'ISO country code':'ISO'}, inplace = True)
df3 = df2.drop('Country', 1)
df4 = pd.merge(df1, df3, on='ISO', how='left')
df4.sort_values(by = 'ISO')
```

```python
df4.isnull().any().sum()
```

```python
df4.to_csv('answer5.csv')
```

### 6. Identify the Data which is present in Data 1 but Missing in Data2.

```python
list_1 = df1.columns
list_2 = df2.columns
name = list(set(list_1) - set(list_2))
print ('Data that is present in Data1 and not present in Data2')
df6 = df1.filter (name)
df6
```

```python
df6.to_csv('answer6.csv')
```

### 7. Get the average of Intensity of deprivation Regional from Data2 for Country-Wise and World Region Wise-Accordingly.


```python
print('Internsity of deprivation regional world region wise')
df2.groupby (['World region'])['Intensity of deprivation Regional'].mean()
```

```python
print('Internsity of deprivation regional Country wise')
df2.groupby (['Country'])['Intensity of deprivation Regional'].mean()
```

```python
writer = pd.ExcelWriter('answer7.xlsx')
df2.groupby (['World region'])['Intensity of deprivation Regional'].mean().to_excel(writer,'Sheet1')
df2.groupby (['Country'])['Intensity of deprivation Regional'].mean().to_excel(writer,'Sheet2')
writer.save()
```
